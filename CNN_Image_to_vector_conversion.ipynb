# Image Processing

```
# Importing Required Libraries
import pandas as pd 
import numpy as np
from IPython.display import display
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
```

### Data Preprocessing

```
from keras.preprocessing.image import ImageDataGenerator
```

#### Preprocessing the Training set

```
train_datagen = ImageDataGenerator(rescale= 1./255)
```

```
train_data = train_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/Saree_Image',
                                               target_size = (256,256),
                                               batch_size = 32)
```

## CNN Pre-trained model

```
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras import layers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization, Activation, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Sequential, Model

frozen = ResNet50(weights="imagenet", input_shape=(256,256,3), include_top=False)
trainable = GlobalAveragePooling2D()(frozen.output)
model = Model(inputs=frozen.input, outputs=trainable)
```

### Taking the images and assigned weight using Resnet

```
from tqdm import tqdm
def get_deep_features(data,model):
    feats=[]
    label=[]
    count=0
    for i in tqdm(data):
        count+=1
        if (count<= 492): # Total Image devide into batch size
            for j in tqdm(i[0]):
              x_in = np.expand_dims(j, axis=0)
              x_in = preprocess_input(x_in)
              features = model.predict(x_in)
              features = features.flatten()
              feats.append(features)
        else: 
            print('dnuogfwif')
            break      

    return feats
```

## Assigning the dataset and model to convert in a vector form

```
vec = get_deep_features( train_data,model)
```
